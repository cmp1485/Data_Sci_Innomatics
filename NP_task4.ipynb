```python
slen = iris[1:,0]
```

```python
swid = iris[1:,1]
```

```python
plen = iris[1:,2]
```

```python
pwid = iris[1:,3]
```

#### Replacing missing values

```python
slen = np.where(np.isnan(slen), np.nanmedian(slen), slen)
```

```python
swid = np.where(np.isnan(swid), np.nanmedian(swid), swid)
```

```python
plen = np.where(np.isnan(plen), np.nanmedian(plen), plen)
```

```python
pwid = np.where(np.isnan(pwid), np.nanmedian(pwid), pwid)
```

#### Outliers

```python
q1slen = np.quantile(slen, 0.25)
q3slen = np.quantile(slen, 0.75)
iqrslen = q3slen - q1slen
iqrslen
```

```python
llslen = q1slen - 1.5*iqrslen
ulslen = q3slen + 1.5*iqrslen
```

```python
np.median(slen)
```

```python
slen_o = (slen<llslen) | (slen>ulslen)
slen[slen_o]
```

```python
slen = np.where(slen_o, np.median(slen), slen)
slen
```

```python
slen_o = (slen<llslen) | (slen>ulslen)
slen[slen_o]
```

```python
q1swid = np.quantile(swid, 0.25)
q3swid = np.quantile(swid, 0.75)
iqrswid = q3swid - q1swid
iqrswid
```

```python
llswid = q1swid - 1.5*iqrswid
ulswid = q3swid + 1.5*iqrswid
```

```python
swid_o = (swid<llswid) | (swid>ulswid)
swid[swid_o]
```

```python
swid = np.where(swid_o, np.median(swid), swid)
swid
```

```python
swid_o = (swid<llswid) | (swid>ulswid)
swid[swid_o]
```

```python
q1plen = np.quantile(plen, 0.25)
q3plen = np.quantile(plen, 0.75)
iqrplen = q3plen - q1plen
iqrplen
```

```python
llplen = q1plen - 1.5*iqrplen
ulplen = q3plen + 1.5*iqrplen
```

```python
plen_o = (plen<llplen) | (plen>ulplen)
plen[plen_o]
```

```python
plen = np.where(plen_o, np.median(plen),plen)
plen
```

```python
plen_o = (plen<llplen) | (plen>ulplen)
plen[plen_o]
```

```python
q1pwid = np.nanquantile(pwid, 0.25)
q3pwid = np.nanquantile(pwid, 0.75)
iqrpwid = q3pwid - q1pwid
iqrpwid
```

```python
llpwid = q1pwid - 1.5*iqrpwid
ulpwid = q3pwid + 1.5*iqrpwid
```

```python
pwid_o = (pwid<llpwid) | (pwid>ulpwid)
pwid[pwid_o]
```

```python
pwid = np.where(pwid_o, np.median(pwid), pwid)
pwid
```

```python
pwid_o = (pwid<llpwid) | (pwid>ulpwid)
pwid[pwid_o]
```
