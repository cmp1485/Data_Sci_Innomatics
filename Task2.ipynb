### Saving the images locally

### Images to arrays

```python
import os
```

```python
arr = []
label = []

for i in lst:
    os.chdir(r"C:\Users\91852\Downloads\DataScience Innomatics Tasks\ML\{}".format(i))

    for j in os.listdir():
        path = os.getcwd() + f'\\{j}'
        img = cv2.imread(path)
        image = cv2.resize(img, (5,6))
        arr.append(image.flatten())
        label.append(i)    
```

```python
df = pd.DataFrame(arr)
```

```python
df['Label']=label
```
